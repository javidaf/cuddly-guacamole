**Show that:**

$$E[(\vec{y} - \tilde{\vec{y}})^2] = \sigma^2 + \text{Bias}[\tilde{\vec{y}}]^2 + \text{Var}[\tilde{\vec{y}}] \tag{4}$$


**Derivation**:
$$E[(\vec{y} - \tilde{\vec{y}})^2] = E[(\vec{y}-f(\vec{x}) + f(\vec{x}) - \tilde{\vec{y}}+E[\tilde{\vec{y}}] - E[\tilde{\vec{y}}])^2]$$

$$= E[((\vec{y}-f(\vec{x}))+(f(\vec{x})-E[\tilde{\vec{y}}])-(\tilde{\vec{y}}-E[\tilde{\vec{y}}]))^2]$$

$$ = E[(\vec{y}-f(\vec{x}))^2+(f(\vec{x})-E[\tilde{\vec{y}}])^2+(\tilde{\vec{y}}-E[\tilde{\vec{y}}])^2 $$

$$+2((\vec{y}-f(\vec{x}))(f(\vec{x})-E[\tilde{\vec{y}}])-(\vec{y}-f(\vec{x}))(\tilde{\vec{y}}-E[\tilde{\vec{y}}])+(f(\vec{x})-E[\tilde{\vec{y}}])(\tilde{\vec{y}}-E[\tilde{\vec{y}}]))]$$

We can now rewrite the above expression as:

$$E[(\vec{y} - \tilde{\vec{y}})^2] = E[(\vec{y}-f(\vec{x}))^2] + E[(f(\vec{x})-E[\tilde{\vec{y}}])^2] + E[(\tilde{\vec{y}}-E[\tilde{\vec{y}}])^2]$$

$$+2E[(\vec{y}-f(\vec{x}))(f(\vec{x})-E[\tilde{\vec{y}}])]-2E[(\vec{y}-f(\vec{x}))(\tilde{\vec{y}}-E[\tilde{\vec{y}}])]+2E[(f(\vec{x})-E[\tilde{\vec{y}}])(\tilde{\vec{y}}-E[\tilde{\vec{y}}])]$$

The first term is the variance of the error:

$$E[(\vec{y}-f(\vec{x}))^2] =E[(f(\vec{x})+\vec{\epsilon}-f(\vec{x}))^2] =E[\vec{\epsilon}^2]= \sigma^2$$

The second term is the bias term:

$$E[(f(\vec{x})-E[\tilde{\vec{y}}])^2] = \text{Bias}[\tilde{\vec{y}}]^2$$

The third term is the variance term:

$$E[(\tilde{\vec{y}}-E[\tilde{\vec{y}}])^2] = \text{Var}[\tilde{\vec{y}}]$$

The fourth term is the covariance between the noise and the model:

$$E[(\vec{y}-f(\vec{x}))(f(\vec{x})-E[\tilde{\vec{y}}])] = 0$$

The fifth term is the covariance between the noise and the deviation from the mean of the model:

$$E[(\vec{y}-f(\vec{x}))(\tilde{\vec{y}}-E[\tilde{\vec{y}}])] = 0$$

The sixth term is the covariance between the deviation from the mean of the model and the model:

$$E[(f(\vec{x})-E[\tilde{\vec{y}}])(\tilde{\vec{y}}-E[\tilde{\vec{y}}])] = 0$$

Thus we have shown that:

$$E[(\vec{y} - \tilde{\vec{y}})^2] = \sigma^2 + \text{Bias}[\tilde{\vec{y}}]^2 + \text{Var}[\tilde{\vec{y}}],\square$$
